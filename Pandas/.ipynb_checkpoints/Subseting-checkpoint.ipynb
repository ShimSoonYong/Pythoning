# Filtering rows from a DataFrame

```python
above_35=titanic[titanic["Age"]>35]
above_35.head()
```

```python
titanic.Age >35
```

__The output of the conditional expression is actually a pandas.Series of boolean values with the same order as the original pandas.DataFrame.__ <br>
<font color="red">Only rows for which the value is _True_ will be selected.</font>

```python
above_35.shape
```

```python
class_23=titanic[titanic["Pclass"].isin([2,3])]
class_23.head()
```

```python
class_23=titanic[(titanic["Pclass"]==2)|(titanic["Pclass"]==3)]
class_23.head()
```

```python
age_no_na=titanic[titanic["Age"].notna()]
age_no_na.head()
```

```python
age_no_na.shape
```
