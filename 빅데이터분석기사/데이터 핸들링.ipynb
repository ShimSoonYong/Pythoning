## 단순 연결

```python
import pandas as pd

HR1 = pd.DataFrame({
    "이름":["장화","홍련"],
    "부서":["영업","회계"],
    "직급":["팀장","사원"]})
HR2=pd.DataFrame({
    "이름":["콩쥐","팥쥐"],
    "직급":["사원","팀장"],
    "부서":["영업","인사"]})
pd.concat([HR1,HR2], axis=0)
```

```python
pd.concat([HR1,HR2], axis=0, ignore_index=True)
```

```python
HR3=pd.DataFrame({
    "이름":["콩쥐","팥쥐"],
    "부서":["영업","인사"],
    "급여":[3500,2800]})
pd.concat([HR1,HR3], axis=0, ignore_index=True)
```

```python
HR4=pd.Series({1:2500}, name="급여")
pd.concat([HR1,HR4], axis=1)
```

```python
HR5=pd.DataFrame(
    {"급여":[4500,3000,3500]})
pd.concat([HR1,HR5],axis=1)
```

## 조인

```python
product=pd.DataFrame({
    "상품코드":["G1","G2","G3","G4"],
    "상품명":["우유","감자","빵","치킨"]})
sale=pd.DataFrame({
    "주문번호":[1001,1002,1002,1003,1004],
    "상품코드":["G4","G3","G1","G3","G5"],
    "주문수량":[1,4,2,2,3]})
sale.merge(product, on="상품코드", how="inner")
```

```python
sale.merge(product, on="상품코드", how="outer", sort=True)
```

```python
sale.merge(product,left_on="상품코드",right_on="상품코드",how="left")
```

```python
sale.merge(product,left_on="상품코드",right_on="상품코드",how="right")
```

# 데이터 요약

## 그룹화와 집계

```python
import pandas as pd
from sklearn.datasets import load_iris

IRIS=load_iris()

iris=pd.DataFrame(data=IRIS.data, columns=IRIS.feature_names)
iris["class"]=IRIS.target
iris["class"]=iris["class"].map({0:"setosa",1:"versicolor",2:"virginica"})

iris.groupby(by="class").mean()
```

```python
iris.groupby("class").median()
```

## 도수분포표

```python
pd.Series(iris["class"]).value_counts().to_frame()
```

```python
iris["petal width level"]=pd.qcut(iris["petal width (cm)"], q=3,
                                  labels=["short","middle","long"])
pd.Series(iris["petal width level"]).value_counts().to_frame()
```

```python
pd.crosstab(iris["petal width level"], iris["class"])
```

# 데이터 재구조화

```python
import pandas as pd

score={"학년":[1,1,1,1,2,2],
       "반":["A","A","B","B","C","C"],
       "성별":["여자","남자","여자","남자","여자","남자"],
       "성적":[76,88,85,72,68,70]}
score=pd.DataFrame(score)
score
```

```python
score=score.pivot_table(index=["학년","반"],
                  columns="성별",values="성적")
score
```

```python
score.reset_index().melt(id_vars=["학년","반"], var_name="성별",
                         value_name="성적")
```

# FUN on DataFrame

```python
import numpy as np
score.apply(np.sqrt, axis=0)
```

```python
score.apply(np.max, axis=0)
```

```python
def plus_five(values):
    return values+5
score.apply(plus_five)
```

```python
def class_avg(df):
    return np.ceil(
        (df["남자"]+df["여자"]) / 2
    )
score.apply(class_avg, axis=1)
```
