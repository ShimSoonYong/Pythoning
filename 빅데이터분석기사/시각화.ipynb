```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.preprocessing import LabelEncoder

iris=load_iris()
iris=pd.DataFrame(iris.data, columns=iris.feature_names)
iris["Class"]=load_iris().target
iris.Class=iris.Class.map({
    0:"Setosa",1:"Versicolour",2:"Virginica"})
```

```python
sns.pairplot(iris, diag_kind="auto", hue="Class")
plt.show()
```

```python
iris_corr=iris.drop(columns="Class",axis=1).corr(method="pearson")
sns.heatmap(iris_corr,
            xticklabels=iris_corr.columns, cmap="RdBu_r", annot=True)
plt.show()
```
